## Naive Bayes Classifier
* basisnya profitabilitas
* asumsi variabel2 independen
* tidak peduli multikolinearity

* h : hipotesis
* d : data

```python
#Feature 1
weather = ['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy','Overcast','Sunny','Sunny','Rainy',
          'Sunny','Overcast','Overcast','Rainy']
# Feature 2
temp = ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild','Cool',
       'Mild','Mild','Mild','Hot','Mild']

# Label atau target
play = ['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes','Yes','Yes','Yes','No']

print(len(weather),len(temp),len(play))
```

### Encode Features

```python
le = LabelEncoder()

w_enc = le.fit_transform(weather)
t_enc = le.fit_transform(temp)
p_enc = le.fit_transform(play)
```

```python
w_enc
```

```python
t_enc
```

```python
p_enc
```

```python
features = list(zip(w_enc,t_enc))
features
```

```python
np.array(features)
```

```python
np.array(features).shape
```

```python
features
```

### Model Building

```python
from sklearn.naive_bayes import GaussianNB

modelNB = GaussianNB()
```

### Model Fitting

```python
modelNB.fit(features,p_enc)

modelNB.predict([[0,2]]) # 0 : overcast, 2 : Mild
```

```python
modelNB.score(features,p_enc) # akurasi 71%
```

## Decision Tree


### Build Decision Tree in Sklearn

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```python
col_names = ['pregnant','glucose','bp','skin','insulin',
            'bmi','predigree','age','label']
pima = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/diabetes.csv',
                  header=None,names=col_names)
pima.head()
```

```python
pima.drop(0,inplace=True)
```

```python
pima.head()
```

```python
pima.info() #dtype object : belum integer, masih string
```

```python
# melihat nol
pima.isna().sum()
```

```python
# mengubah menjadi float
for i in pima.columns:
    pima[i] = pima[i].astype(float)
```

```python
pima.info()
```

```python
pima.head()
```

```python
feature_cols = ['pregnant','insulin','bmi','age','glucose','bp','predigree']

x = pima[feature_cols]
y = pima['label']
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=2024)
```

```python
# Build Model

dtc = DecisionTreeClassifier()

dtc.fit(X_train,y_train)

y_pred_train = dtc.predict(X_train)
y_pred_test = dtc.predict(X_test)
```

```python
print('train set \n', classification_report(y_train,y_pred_train))
print('test set \n', classification_report(y_test,y_pred_test))
```

model diatas overfitting

### Visualisasikan Decision Tree

```python
dot_data = StringIO()

tree.export_graphviz(dtc, out_file=dot_data,
                    class_names=['0','1'], #nama target
                    feature_names = feature_cols,
                    filled=True,
                    rounded=True,
                    special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

Image(graph.create_png())
```

### Ganti max_dept jadi 3

```python
dtc2 = DecisionTreeClassifier(max_depth=3) #prepruning
dtc2.fit(X_train,y_train)

y_pred_train = dtc2.predict(X_train)
y_pred_test = dtc2.predict(X_test)
```

```python
print('train \n', classification_report(y_train,y_pred_train))
print('test \n', classification_report(y_test,y_pred_test))
```
